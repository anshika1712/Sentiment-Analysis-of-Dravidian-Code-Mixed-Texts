```python
import string
import re
import emoji

#Demojizing
def emoji_tagging(texts):
  result=emoji.demojize(texts, delimiters=(" "," "))

  return result

#removing urls
def remove_urls(texts):
    url_remove = re.compile(r'https?://\S+|www\.\S+')
    return url_remove.sub(r'', texts)

#removing htlm links
def remove_html(texts):
    html=re.compile(r'<.*?>')
    return html.sub(r'',texts)

#Number removal
def remove_num(texts):
    remove= re.sub(r'\d+', '', texts)
    return remove

#removing punctuations
def punct_remove(sentence):
  s={'!', '"', "'", '#', '$', '%', '&', '(', ')', '*', '+', ',', '-', '.', '/', ':', ';', '<', '=', '>', '?', '@', '[', ']', '^', '{', '|', '}', '~','..'}
  sentence = word_tokenize(sentence)
  filtered_word = []

  for i in sentence:
    letter=[]
    for k in i:
      if k not in s:
        letter.append(k);
    listTo= ''.join([str(elem) for elem in letter])
    filtered_word.append(listTo)
  listToStr = ' '.join([str(elem) for elem in filtered_word])
  return(listToStr)

#removing mentions
def remove_mention(x):
    texts=re.sub(r'@\w+','',x)
    return texts

#remove underscore
def underscore(x):
  filteredText=re.sub('_', ' ',x)

  return filteredText

#Removing repeated characters
#identical consecutive letters reduce them to max 2 consecutive occurrences of the same letters
def remove_repeat(texts):
    repeat_char = re.compile(r"(.)\1{1,}", re.IGNORECASE)
    texts = repeat_char.sub(r"\1\1", texts)
    texts = re.sub("(.)\\1{2,}", "\\1", texts)
    return texts

#stripping white spaces
def strip(x):
  texts=x.strip()
  return texts

#removing extra space
def space(x):
  texts=re.sub(' +',' ',x)
  return (texts)
```

```python
%%time
from tqdm.auto import tqdm
data['clean_text']=[space(strip(punct_remove(underscore(remove_mention(remove_repeat(remove_num(remove_html(remove_urls(emoji_tagging(t)))))))))) for t in tqdm(data['text'].values)]
```
