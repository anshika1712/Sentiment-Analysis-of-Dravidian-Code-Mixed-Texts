```python
dictionary=word_list.set_index('Word').to_dict()['Language']
```

```python
#checking if the word has multiple language tag and replacing it with 'ta'
for comments in test_data['LID list']:
  for words in comments:
    if any(isinstance(i, list) for i in words):
      if 'en' and 'hi' not in words[1]:
        words[1]='kn'
      elif 'hi' in words[1]:
        words[1]='hi'
      elif 'en' in words[1]:
        words[1]='en'
      print(words)
```

```python
#lang=['hi','ml','en','ta','kn','te']
for comments in test_data['LID list']:
  for words in comments:
    if words[0] in dictionary:
      words[1] = dictionary[words[0]]
```

```python
def checking(sen):
  for words in sen:
    if words[1] == 'ta' or words[1] == 'Unknown':
    #if words[1] == 'ml':
      return 'tamil'
    else:
      continue
  return 'not-tamil'
```

```python
%%time
from tqdm.auto import tqdm
train_data['check']=[checking(t) for t in tqdm(train_data['LID list'].values)]
train_data.head()
```

```python
%%time
from tqdm.auto import tqdm
test_data['check']=[checking(t) for t in tqdm(test_data['LID list'].values)]
test_data.head()
```

```python
%%time
from tqdm.auto import tqdm
val_data['check']=[checking(t) for t in tqdm(val_data['LID list'].values)]
val_data.head()
```
