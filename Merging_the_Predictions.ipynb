```python
#Importing the library and loading the dataset
import pandas as pd
import numpy as np
#Reading the dataset
data=pd.read_csv('/content/KA-XLMR_Full_Q3.tsv', sep="\t")
```

```python
data.head()
```

```python
#loading the second dataset
data2=pd.read_csv('/content/Kan_or_notKan_test.csv')
data2.head()
```

```python
#data3=pd.read_csv('/content/correct_Kan_test (1).csv')
#data3.head()
```

```python
#filtering the not in lang values
no=data2.loc[data2['check']=='not-Kannada']
```

```python
no.head()
```

```python
no.drop(['Unnamed: 0',	'category',	'text',	'clean_text',	'tokenized_text',	'LID list'],axis=1, inplace=True)
```

```python
no.head()
```

```python
no.rename(columns = {'check':'category'}, inplace = True)
```

```python
frames=[data, no]
```

```python
result = pd.concat(frames)
```

```python
result.sort_values(by=['id'])
```

```python
#checking for null values
result.isnull().sum()
```

```python
#Function to remove 'Kan_' to get the numbers from the id
def disect(ids):
  ids=ids[4:]
  return ids
```

```python
%%time
from tqdm.auto import tqdm
result['id2']=[disect(t) for t in tqdm(result['id'].values)]
result.head()
```

```python
result.dtypes
```

```python
#typecasting to int in order to sort the values
convert_dict = {'id2': int
                }

result = result.astype(convert_dict)
```

```python
10000//2400
```

```python
result.sort_values(by='id2',inplace=True)
result
```

```python
result.drop(['id2'], axis=1, inplace=True)
result
```

```python
# saving as tsv file
result.to_csv('Kan-XLMR.tsv', sep="\t")
```
